**5. Check and handle outliers in at least 3 columns in the dataset**

```python
columns_to_check = ['Age', 'Fare', 'SibSp']
df.boxplot(columns_to_check)
```

```python
Q1 = df[columns_to_check].quantile(0.25)
Q3 = df[columns_to_check].quantile(0.75)
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = ((df[columns_to_check] < lower_bound) | (df[columns_to_check] > upper_bound)).sum()
print(outliers)

# Replace outliers with mean
for col in columns_to_check:
    mean_val = df[col].mean()
    df[col] = np.where((df[col] < lower_bound[col]) | (df[col] > upper_bound[col]), mean_val, df[col])
```

**6.Do min max scaling on the feature set (Take ‘Survived’ as target)**

```python
from sklearn.preprocessing import MinMaxScaler

# Separate features and target
x = df.drop('Survived', axis=1)
y = df['Survived']

# Apply Min-Max Scaling to the feature set
scaler = MinMaxScaler()
x_scaled = pd.DataFrame(scaler.fit_transform(x), columns=x.columns, index=x.index)
# Combine the scaled features with the target column
df_scaled = pd.concat([x_scaled, y], axis=1)

# Display the first few rows of the scaled dataframe
df_scaled.head()
```
