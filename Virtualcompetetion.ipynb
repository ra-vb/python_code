```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_data=pd.read_csv('/content/train_ctrUa4K.csv')
test_data=pd.read_csv('/content/test_lAUu6dG.csv')
tr_data=train_data.copy()
ts_data=test_data.copy()
```

```python
# Separate numerical and categorical columns
numerical_cols = tr_data.select_dtypes(include=['number']).columns
categorical_cols = tr_data.select_dtypes(include=['object']).columns
numerical_cols
```

```python
categorical_cols
```

```python
for col in numerical_cols:
    tr_data[col] = tr_data[col].fillna(tr_data[col].median())
tr_data.isnull().sum()
```

```python
# prompt: find outliers and fill with mean
#AGE  150, 160

import numpy as np
# Calculate IQR for numerical columns
Q1 = tr_data[numerical_cols].quantile(0.25)
Q3 = tr_data[numerical_cols].quantile(0.75)
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = ((tr_data[numerical_cols] < lower_bound) | (tr_data[numerical_cols] > upper_bound)).sum()
print(outliers)

# Replace outliers with mean
for col in numerical_cols:
    mean_val = tr_data[col].mean()
    tr_data[col] = np.where((tr_data[col] < lower_bound[col]) | (tr_data[col] > upper_bound[col]), mean_val, tr_data[col])
```

```python
tr_data
```

```python
# Normalize numerical features (using MinMaxScaler)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
tr_data[numerical_cols] = scaler.fit_transform(tr_data[numerical_cols])
tr_data
```

```python
# Label encode categorical features
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
for col in categorical_cols:
    tr_data[col] = encoder.fit_transform(tr_data[col])
tr_data
```

```python
y= tr_data['Loan_Status'] #Labels / Target variable
x= tr_data.drop(columns=['Loan_Status'])  #features
```

```python
#train ad test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#model creation
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
print('Accuracy is', accuracy_score(y_test, y_pred))
print('Precision is', precision_score(y_test, y_pred))
print('Recall is', recall_score(y_test, y_pred))
print('f1 score is', f1_score(y_test,y_pred))
```

```python
r2 = lr.score(x_test, y_test)
r2
```

```python
import joblib
joblib.dump(lr, 'loan_model.pkl')
```
