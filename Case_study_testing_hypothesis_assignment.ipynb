```python
import pandas as pd
import numpy as np
import math
from scipy.stats import norm
data=pd.read_csv('/content/Sales_add.csv')
data.info()
```

```python
#H0 mu=x
#H1:mu>x
import scipy.stats as stats

mean_before = 149239.954545
std_before = 14844.042921
size = 22
mean_after = 231123.727273
std_after = 25556.777061
np.random.seed(0)
Sales_before_digital_add = np.random.normal(loc=mean_before, scale=std_before, size=size)
Sales_after_digital_add = np.random.normal(loc=mean_after, scale=std_after, size=size)
t_statistic, p_value = stats.ttest_rel(Sales_after_digital_add, Sales_before_digital_add,alternative='greater')
print(f'T-statistic: {t_statistic}')
print(f'P-value: {p_value}')
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant increase in sales after digital marketing.")
else:
    print("Fail to reject the null hypothesis. There is no significant increase in sales after digital marketing.")
```

```python
from scipy.stats import chi2_contingency
#print(data.head())
contingency_table = pd.crosstab(data['Region'], data['Manager'])
chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)
print(f'Chi-square statistic: {chi2_stat}')
print(f'P-value: {p_val}')
print(f'Degrees of freedom: {dof}')
print('Expected frequencies table:')
print(expected)
alpha=0.05
if p_value<alpha:
  print("Reject the null hypothesis:there is any dependency between the features region and manager")
else:
  print("Fail to rejct the hypothesis: there is no dependency between the features region and manager")
```
